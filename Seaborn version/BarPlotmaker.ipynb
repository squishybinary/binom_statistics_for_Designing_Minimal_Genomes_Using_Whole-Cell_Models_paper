```python
import seaborn as sns
import pandas as pd
import matplotlib
#import matplotlib.pyplot as plt

#Change width of bars
# https://stackoverflow.com/questions/34888058/changing-width-of-bars-in-bar-chart-created-using-seaborn-factorplot
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

#Custom colour palette        
greyscale_colours = ['#d6d6d6',  # Light Grey
                    '#969696',  # Medium Grey
                    '#585757',  # Dark Grey
                   ]
#Bar Chart Data
#d = {'x': ["GAMA_236", "Model"], 'y': [236,401]}
d = {'x': ["Lego_261", "Model"], 'y': [261,401]}
df = pd.DataFrame(data=d)

#Plotting
fig, ax = plt.subplots(figsize=(4,5.5))
plot = sns.barplot(x="x", y="y", data=df,  palette=greyscale_colours)

#Settings
ax.set_ylabel("in-silico genome size (genes)")
ax.set_xlabel('')
plt.setp(fig.axes, yticks=[]) #remove axis ticks
change_width(ax, .3) #change width of bars
#sns.despine() #remove right and top lines of chart
sns.despine(left=True) # remove right, top, and left lines of chart
plt.style.use(['seaborn-white', 'seaborn-talk']) #set styles and relative object sizes
#plt.style.use(['seaborn-white', 'seaborn-paper'])
#plt.style.use(['seaborn-white', 'seaborn-poster'])

#Export figure
figure = plot.get_figure()    
#figure.savefig('GAMAbarplot.png', dpi=400)
#figure.savefig('GAMAbarplot.svg', dpi=400)
figure.savefig('Legobarplot.png', dpi=400)
figure.savefig('Legobarplot.svg', dpi=400)
```
